*** Performing variable level analysis


```python
import matplotlib.pyplot as plt
```

```python
def box_plot(df, col):
    df = df.copy()
    plt.boxplot(df[col])
    plt.title(col)
    plt.show()
    
for col in ingredients.columns:
    box_plot(ingredients, col)
```

```python
def hist_plot(df, col):
    df = df.copy()
    df[col].hist(bins=20)
    plt.ylabel('cnt')
    plt.xlabel(col)
    plt.title(col)
    plt.show()
    
for col in ingredients.columns:
    hist_plot(ingredients, col)
```

### Summary of observations: 
- From the Histogram - f,h,i are having skewed distribution. 
- Positively corelated: a & g  
- Negatively corelated: e & a 

#### Kmeans and hierarchical clustering 

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler
```

```python
X = ingredients.copy()
```

```python
SS = []
for cluster in range(1, 11):
    kmeans = KMeans(n_clusters=cluster, init='k-means++', max_iter=300, n_init=10, random_state=0)
    kmeans.fit(X)
    SS.append(kmeans.inertia_)
plt.plot(range(1, 11), SS)
plt.title('Elbow Plot')
plt.xlabel('Number of clusters')
plt.ylabel('SS')
plt.show()
```

```python
import scipy.cluster.hierarchy as sch
dendrogram = sch.dendrogram(sch.linkage(X, method = 'ward'))
plt.title('Dendograms')
plt.xlabel('Ingredients')
plt.ylabel('Euclidean distances')
plt.show()
```

Observations:
- As per both Elbow and hierarchical analysis (high level), there are 3 clusters
